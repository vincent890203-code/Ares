# 1. 專案初始化 (只做一次)
```Bash
# 初始化 Git 倉庫 (讓 Git 開始監控這個資料夾)
git init

# (選用) 建立 .gitignore 檔案，忽略不必要的檔案 (如 __pycache__, .DS_Store)
# Windows PowerShell 指令:
echo "__pycache__/" > .gitignore
echo "*.pyc" >> .gitignore
echo ".pytest_cache/" >> .gitignore
echo "brain_memory/" >> .gitignore  # 通常我們不傳大模型到 git，除非很小
```

# 2. 日常開發循環 (The Daily Loop)
```Bash
# 步驟 1: 檢查狀態 (最重要的指令！)
# 告訴您哪些檔案被修改了、哪些是新檔案
git status

# 步驟 2: 將檔案加入「暫存區」 (準備要存檔的檔案)
# "." 代表加入所有變更
git add .

# 步驟 3: 提交版本 (正式存檔)
# -m 後面接的是我們 DEV_LOG 裡的標題，例如：
git commit -m "feat: Implement MVC architecture and Registry module (v0.4.0)"

```

# 3. 查看歷史紀錄 (Time Machine)
```Bash
# 查看詳細歷史 (包含作者、時間、Hash)
git log

# 查看簡潔歷史 (一行一個版本，很適合快速瀏覽)
git log --oneline
```

# 4. 這些不是 Git，但我們這幾天一直在用的指令
```Bash

# 1. 編輯模式安裝套件 (Day 1 重點)
# 讓您修改程式碼後不用重新安裝就能生效
pip install -e .

# 2. 執行單元測試 (Day 1-2 重點)
pytest

# 3. 執行我們寫的整合測試腳本 (Day 2-3 重點)
python test_integration_v2.py

# 4. 清理測試產生的垃圾 (Windows PowerShell)
# 當我們想重置測試環境時
Remove-Item -Recurse -Force brain_memory_test

```
